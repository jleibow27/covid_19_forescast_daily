# COVID-19 Time Series Prediction - Confirmed Cases
***
## Import Packages

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split

%matplotlib inline

# This will allow us to avoid a FutureWarning when plotting.
from pandas.plotting import register_matplotlib_converters
register_matplotlib_converters()
```

## Read in Training data

```python
cov = pd.read_csv('../data/train.csv')
```

## Set index to Datetime

```python
cov.set_index('Date', inplace=True)
```

```python
cov.sort_index(inplace=True)
```

```python
cov
```

```python
cov['Country/Region'].value_counts()
```

## Select Only Confirmed Cases

```python
cov_confirmed = cov.groupby('Date')[['ConfirmedCases']].sum()
```

```python
cov_confirmed.tail()
```

## Set a rolling mean of 3

```python
cov_confirmed.rolling(3).mean()
```

## Create Lag Columns

```python
# Create a column called `lag_1` that lags Confirmed Cases by 1 day
cov_confirmed['lag_1'] = cov_confirmed['ConfirmedCases'].shift(1)

# Create a column called `lag_2` that lags Confirmed Cases by 2 days.
cov_confirmed['lag_2'] = cov_confirmed['ConfirmedCases'].shift(2)

# Create a column called `seasonal_12` that lags Confirmed Cases by 4 days.
cov_confirmed['lag_4'] = cov_confirmed['ConfirmedCases'].shift(4)

# Create a variable called `time` that takes on a value of 0 on January 22nd 2020,
# then increases by 1 each day until the end of the dataframe.
cov_confirmed['time'] = range(0, cov_confirmed.shape[0])
```

```python
cov_confirmed.head()
```

```python
cov_confirmed.tail()
```

## Create a Model by Generating train/test split

```python
X_train, X_test, y_train, y_test = train_test_split(cov_confirmed.drop(columns='ConfirmedCases'),
                                                    cov_confirmed['ConfirmedCases'],
                                                    test_size = 0.1,
                                                    shuffle = False)
```

```python
# Check shape to confirm this was done properly.
print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
# Import statsmodels.
import statsmodels.api as sm
```

## Create an intercept as a column entitled "constant"

```python
# Create intercept

X_train = sm.add_constant(X_train)
X_test = sm.add_constant(X_test)

# Confirm.
X_train.head()
```

## Drop the missing values

```python
# statsmodels won't be able to handle missing values.

X_train.dropna(inplace = True)
y_train = y_train[X_train.index]

# This way we subset y_train to keep only indices from X_train.
```

## Instantiate model

```python
# In statsmodels, we pass our data in when we instantiate the model!

lm = sm.OLS(y_train, X_train)
```

## Fit model

```python
# Fit the model.
lm_results = lm.fit()
```

## Display statsmodels summary

```python
print(lm_results.summary())
```

```python
# Generate predicted test values.
lm_results.predict(X_test)
```

```python
# Import R2 score and MSE.
from sklearn.metrics import r2_score, mean_squared_error
```

## Calculate R2 and Root Mean Squared Error (RMSE) Scores

```python
print(f'R2: {r2_score(y_test, lm_results.predict(X_test))}')
print(f'RMSE: {mean_squared_error(y_test, lm_results.predict(X_test)) ** 0.5}')
```

## Plot predictions! 

```python
# Set figure size.
plt.figure(figsize=(20,10))

# Plot training data.
plt.plot(y_train.index, y_train.values, color = 'blue')

# Plot testing data.
plt.plot(y_test.index, y_test.values, color = 'orange')

# Plot predicted test values.
plt.plot(lm_results.predict(X_test), color = 'green')

# Set label.
plt.title(label = 'Forecasting COVID-19 cases Worldwide', fontsize=24)

# Resize tick marks.
plt.xticks(fontsize=14, rotation=80)
plt.yticks(fontsize=20);
```

## Predicting Fatalities

```python
cov_fatal = cov.groupby('Date')[['Fatalities']].sum()
```

```python
cov_fatal.tail()
```

## Create a Rolling Mean of 3

```python
cov_fatal.rolling(3).mean()
```

## Create Lag Columns

```python
# Create a column called `lag_1` that lags Fatality Cases by 1 day
cov_fatal['lag_1'] = cov_fatal['Fatalities'].shift(1)

# Create a column called `lag_2` that lags Fatality Cases by 2 days.
cov_fatal['lag_2'] = cov_fatal['Fatalities'].shift(2)

# Create a column called `seasonal_12` that lags Fatality Cases by 4 days.
cov_fatal['lag_4'] = cov_fatal['Fatalities'].shift(4)

# Create a variable called `time` that takes on a value of 0 on January 22nd 2020,
# then increases by 1 each day until the end of the dataframe.
cov_fatal['time'] = range(0, cov_fatal.shape[0])
```

## Check shape to confirm this was done properly.

```python
# Check shape to confirm this was done properly.
print(X_train_fa.shape)
print(X_test_fa.shape)
print(y_train_fa.shape)
print(y_test_fa.shape)
```

## Create an intercept as a column entitled "constant"

```python
# Create intercept

X_train_fa = sm.add_constant(X_train_fa)
X_test_fa = sm.add_constant(X_test_fa)

# Confirm.
X_train_fa.head()
```

```python
# statsmodels won't be able to handle missing values.

X_train_fa.dropna(inplace = True)
y_train_fa = y_train_fa[X_train_fa.index]

# This way we subset y_train to keep only indices from X_train.
```

```python
# In statsmodels, we pass our data in when we instantiate the model!

lm_fa = sm.OLS(y_train_fa, X_train_fa)
```

## Fit Model

```python
# Fit the model.
lm_fa_results = lm_fa.fit()
```

## Print a Statsmodels summary

```python
print(lm_fa_results.summary())
```

## Generate predicted test values.

```python
lm_fa_results.predict(X_test_fa)
```

## Print R2 and RMSE scores

```python
print(f'R2: {r2_score(y_test_fa, lm_fa_results.predict(X_test_fa))}')
print(f'RMSE: {mean_squared_error(y_test_fa, lm_fa_results.predict(X_test_fa)) ** 0.5}')
```
