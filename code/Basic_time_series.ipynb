```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split

%matplotlib inline

# This will allow us to avoid a FutureWarning when plotting.
from pandas.plotting import register_matplotlib_converters
register_matplotlib_converters()
```

```python
cov = pd.read_csv('../data/train.csv')
```

```python
cov.set_index('Date', inplace=True)
```

```python
cov.sort_index(inplace=True)
```

```python
cov
```

```python
cov_us = cov.loc[cov['Country/Region'] == 'US']
```

```python
cov_us.head(5)
```

```python
cov_us_confirmed = cov_us.groupby('Date')[['ConfirmedCases']].sum()
```

```python
cov_us_confirmed.tail()
```

```python
cov_us_confirmed.rolling(3).mean()
```

```python
# Create a column called `lag_1` that lags Confirmed Cases by 1 day
cov_us_confirmed['lag_1'] = cov_us_confirmed['ConfirmedCases'].shift(1)

# Create a column called `lag_2` that lags Confirmed Cases by 2 days.
cov_us_confirmed['lag_2'] = cov_us_confirmed['ConfirmedCases'].shift(2)

# Create a column called `seasonal_12` that lags Confirmed Cases by 4 days.
cov_us_confirmed['lag_4'] = cov_us_confirmed['ConfirmedCases'].shift(4)

# Create a variable called `time` that takes on a value of 0 on January 22nd 2020,
# then increases by 1 each day until the end of the dataframe.
cov_us_confirmed['time'] = range(0, cov_us_confirmed.shape[0])
```

```python
cov_us_confirmed.head()
```

```python
cov_us_confirmed.tail()
```

```python
# Generate train/test split.
X_train, X_test, y_train, y_test = train_test_split(cov_us_confirmed.drop(columns='ConfirmedCases'),
                                                    cov_us_confirmed['ConfirmedCases'],
                                                    test_size = 0.1,
                                                    shuffle = False)
```

```python
# Check shape to confirm this was done properly.
print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
# Import statsmodels.
import statsmodels.api as sm
```

```python
# Before fitting a model in statsmodels, what do we need
# to do? (Hint: Think intercept.)

X_train = sm.add_constant(X_train)
X_test = sm.add_constant(X_test)

# Confirm.
X_train.head()
```

```python
# statsmodels won't be able to handle missing values.

X_train.dropna(inplace = True)
y_train = y_train[X_train.index]

# This way we subset y_train to keep only indices from X_train.
```

```python
# In statsmodels, we pass our data in when we instantiate the model!

lm = sm.OLS(y_train, X_train)
```

```python
# Fit the model.
lm_results = lm.fit()
```

```python
# Display a summary!
print(lm_results.summary())
```

```python
# Generate predicted test values.
lm_results.predict(X_test)
```

```python
# Import R2 score and MSE.
from sklearn.metrics import r2_score, mean_squared_error
```

```python
# Calculate R2 score.
r2_score(y_test, lm_results.predict(X_test))
```

```python
# Calculate RMSE.
mean_squared_error(y_test, lm_results.predict(X_test)) ** 0.5
```

```python
# Plot predictions! 

# Set figure size.
plt.figure(figsize=(20,10))

# Plot training data.
plt.plot(y_train.index, y_train.values, color = 'blue')

# Plot testing data.
plt.plot(y_test.index, y_test.values, color = 'orange')

# Plot predicted test values.
plt.plot(lm_results.predict(X_test), color = 'green')

# Set label.
plt.title(label = 'Forecasting COVID-19 cases in the US', fontsize=24)

# Resize tick marks.
plt.xticks(fontsize=20)
plt.yticks(fontsize=20);
```
